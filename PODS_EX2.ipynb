```python
import pandas as pd
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
a=pd.read_csv("Titanic-Dataset.csv")
df=pd.DataFrame(a)
print(df)
```

```python
df.shape
```

```python
imputer = SimpleImputer(strategy='mean')
df['Age'] = imputer.fit_transform(df[['Age']])
df
```

```python
df['Cabin']=df['Cabin'].fillna("Unknown")
df
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

```python
import seaborn as sns
sns.countplot(x='Pclass', data=df)
plt.title('Distribution of Passengers by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
female_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print("Names of female passengers who survived:")
for name in female_survived['Name']:
    print(name)
```

```python
third_class_under_18 = df[(df['Pclass'] == 3) & (df['Age'] < 18)]
display(third_class_under_18)
```

```python
first_class_above_40 = df[(df['Pclass'] == 1) & (df['Age'] > 40)]
display(first_class_above_40)
```

```python
survived = df[(df['Survived'] == 1)]
print("Names of fepassengers who survived:")
for name in survived['Name']:
    print(name)
```

```python
fare_greater_than_100 = df[ (df['Fare'] > 100) & (df['Sex'] == 'male')]
display(fare_greater_than_100)
```

```python
a = df[(df['Embarked'] == 'C') & (df['Pclass'] == 2)]
display(a)
```

```python
a = df[(df['SibSp'] == 0) & (df['Parch'] == 0)]
display(a)
```

```python
oldest_survived = df[df['Survived'] == 1].sort_values(by='Age', ascending=False).head(5)
display(oldest_survived[['Name', 'Age']])
```

```python
a = df[(df['Fare'] == 0)]
display(a)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print("Training Dataset:")
print(train_df.head(3))
print("\nTesting Dataset:")
print(test_df.head(3))
```
